## Data Cleaning

```python
# We'll first drop useless columns so, they won't make a problem in the process of modeling

breast_cancer_df.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
breast_cancer_df.head()
```

```python
# Transform target variable to be numerical because we can't feed our models non-numerical data

diagnosis_dict = {'B' : 0, 'M' : 1}        ### B = Benign ---> 0 / M = Malignant ---> 1
breast_cancer_df['diagnosis'].replace(diagnosis_dict, inplace=True)
breast_cancer_df['diagnosis'].value_counts()
```

## Correlation

```python
# Split the list of predictors into 3 different lists (mean, se & worst)

mean_predictors = list(breast_cancer_df.columns[1:11])
mean_predictors.append('diagnosis')
se_predictors = list(breast_cancer_df.columns[11:21])
se_predictors.append('diagnosis')
worst_predictors = list(breast_cancer_df.columns[21:])
worst_predictors.append('diagnosis')
print('The mean predictors are: ', mean_predictors)
print('The se predictors are: ', se_predictors)
print('The worst predictors are:', worst_predictors)
```

```python
# Calculate the correlation coefficient between target & 'mean' predictors

mean_corr = breast_cancer_df[mean_predictors].corr()
sns.heatmap(mean_corr, annot=True)
plt.title('Correlation between target and mean predictors')
plt.show()
```

```python
# Calculate the correlation coefficient between target & 'se' predictors

mean_corr = breast_cancer_df[se_predictors].corr()
sns.heatmap(mean_corr, annot=True)
plt.title('Correlation between target and se predictors')
plt.show()
```

```python
# Calculate the correlation coefficient between target & 'worst' predictors

mean_corr = breast_cancer_df[worst_predictors].corr()
sns.heatmap(mean_corr, annot=True)
plt.title('Correlation between target and worst predictors')
plt.show()
```

# Modeling :

```python
# Extracting the features and labels

X = breast_cancer_df.drop('diagnosis', axis=1)
y = breast_cancer_df['diagnosis']
```

```python
# Splitting data into training and test datasets

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Standardize the data using a standard scaler to take care of outliers

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_1 = scaler.fit_transform(X_train)
X_test_1 = scaler.transform(X_test)
```

```python
# Building up the layers of the neural network model

import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# Compiling our newly-made model

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Training our model on training data

history = model.fit(X_train_1, y_train, epochs=10, validation_split=0.1)
```

```python
# Visualizing training and test accuracy according to the number of epochs

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
# Visualizing training and test loss according to the number of epochs

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
# Evaluate our model on test data

test_loss, test_acc = model.evaluate(X_test_1, y_test)
print('Test accuracy:', test_acc)
```

```python
# Make predictions on test data and transform them into class labels

y_pred = model.predict(X_test_1)
y_pred_labels = np.argmax(y_pred, axis=1)
y_pred_labels
```
